```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/diogenesjusto/FIAP-On-Line-Graduation/main/datasets/arvore_titanic.csv')
df.head(13)
```

```python
from pycaret.classification import *
s = setup(df, target = 'Survived')
```

```python
best = compare_models()
```

```python
print(best)
```

```python
evaluate_model(best)
```

```python
modelos_5 = compare_models(n_select = 5)
```
